### YFinance Pull

```python
start="2020-01-01"

end="2024-12-31"
```

```python

all_data = {}            # stores sucessful tickers 
error_tickers = {}       # stores failed tickers

for ticker in ls_ticker:
    try:
        print(f"Downloading: {ticker}")
        data = yf.download(ticker, start=start, end=end)

        # Empty data means invalid ticker or no data
        if data.empty:
            print(f" No data returned for {ticker}. Skipping.")
            error_tickers[ticker] = "No data returned"
            continue
        
        # Save valid data
        all_data[ticker] = data

    except Exception as e:
        print(f"!!Error downloading {ticker}: {e}")
        error_tickers[ticker] = str(e)
        continue


# Combine all downloaded tickers into one multi-index dataframe

if all_data:
    final_df = pd.concat(all_data, axis=1)
else:
    final_df = pd.DataFrame()

# Show results
print("\n Final combined DataFrame:")
print(final_df.head())

print("\n!!! Error tickers dictionary:")
print(error_tickers)
```
